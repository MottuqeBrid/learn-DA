## 7.2 Data Transformation

### Removing Duplicates

```python
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"],
                     "k2": [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
data.duplicated()  # Check for duplicate rows
```

```python
data.drop_duplicates()  # Drop duplicate rows
```

```python
data
```

```python
data["v1"] = range(7)
data
```

```python
data.drop_duplicates(subset=["k1"])  # Drop duplicates based on column 'k1'
```

```python
data.drop_duplicates(["k1", "k2"], keep="last")  # Drop duplicates based on columns 'k1' and 'k2'
```

### Transforming Data Using a Function or Mapping

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
 "bacon": "pig",
 "pulled pork": "pig",
 "pastrami": "cow",
 "corned beef": "cow",
 "honey ham": "pig",
 "nova lox": "salmon"
}
data["animal"] = data["food"].map(meat_to_animal)
data
```

```python
def get_animal(x):
    return meat_to_animal[x]
data["food"].map(get_animal)
```

### Replacing Values

```python
data =pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
data.replace(-999, np.nan)  # Replace -999 with NaN
```

```python
data.replace([-999, -1000], np.nan)  # Replace -999 and -1000 with NaN
```

```python
data.replace([-999, -1000], [np.nan, 0])  # Replace -999 with NaN and -1000 with 0
```

```python
data.replace({-999:np.nan, -1000:0})  # Replace -999 with NaN and -1000 with 0 using a dictionary
```

### Renaming Axis Indexes

```python
data = pd.DataFrame(np.arange(12).reshape(3, 4), 
                    index=["Ohio", "Colorado", "New York"],
                    columns=["one", "two", "three", "four"])
data
```

```python
def transform(x):
    return x[:4].upper()
data.index = data.index.map(transform)
```

```python
data
```

```python
data.rename(index=str.title, columns=str.upper)  # Rename index and columns with functions
```

```python
data
```

```python
data.rename(index={"OHIO": "INDIANA"}, columns={"three": "peekaboo"})  # Rename specific index and column labels
```
