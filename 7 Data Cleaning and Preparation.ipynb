## 7.2 Data Transformation

### Removing Duplicates

```python
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"],
                     "k2": [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
data.duplicated()  # Check for duplicate rows
```

```python
data.drop_duplicates()  # Drop duplicate rows
```

```python
data
```

```python
data["v1"] = range(7)
data
```

```python
data.drop_duplicates(subset=["k1"])  # Drop duplicates based on column 'k1'
```

```python
data.drop_duplicates(["k1", "k2"], keep="last")  # Drop duplicates based on columns 'k1' and 'k2'
```

### Transforming Data Using a Function or Mapping

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
 "bacon": "pig",
 "pulled pork": "pig",
 "pastrami": "cow",
 "corned beef": "cow",
 "honey ham": "pig",
 "nova lox": "salmon"
}
data["animal"] = data["food"].map(meat_to_animal)
data
```

```python
def get_animal(x):
    return meat_to_animal[x]
data["food"].map(get_animal)
```

### Discretization and Binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
```

```python
age_categories = pd.cut(ages, bins)
age_categories
```

```python
age_categories.codes
```

```python
age_categories.categories
```

```python
age_categories.categories[0]
```

```python
pd.value_counts(age_categories)
```

```python
pd.cut(ages, bins, right=False)  # Left-inclusive bins
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]
pd.cut(ages, bins, labels=group_names)  # Use custom labels for the bins
```

```python
data = np.random.uniform(size=20)
pd.cut(data, 4, precision=2)  # Cut into 4 equal-width bins with 2 decimal places
```

```python
data = np.random.standard_normal(1000)
quartiles = pd.qcut(data, 4, precision=2)  # Cut into 4 equal-size bins with 2 decimal places
quartiles
```

```python
pd.value_counts(quartiles)
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.0]).value_counts()
```
